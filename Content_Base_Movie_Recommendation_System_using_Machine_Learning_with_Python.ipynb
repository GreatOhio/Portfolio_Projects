```python
import numpy as np
import pandas as pd

import difflib

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies_1.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

#### Feature selection 

```python
movies_data_features = movies_data[["genres","keywords","overview","tagline","cast","director"]]
```

```python
movies_data_features.fillna("", inplace = True)
```

```python
movies_data_features.isnull().sum()
```

```python
combine_movies_data_features = movies_data_features["genres"] + " " + movies_data_features["keywords"] + " " + movies_data_features["overview"] + " " + movies_data_features["tagline"]+ " " + movies_data_features["cast"]+ " " + movies_data_features["director"]
combine_movies_data_features
```

#### Converting combined text features to vectors

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combine_movies_data_features)
```

```python
print(feature_vectors)
```

#### Getting similarity score

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

#### Recommendation system

```python
movie_name = input ('Enter your favorite movie name: ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data['title'] == close_match]["index"].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x: x[1], reverse = True)

print ('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data['index'] == index]["title"].values[0]
    if (i<31):
        print(i, '.', title_from_index)
        i+=1
```
