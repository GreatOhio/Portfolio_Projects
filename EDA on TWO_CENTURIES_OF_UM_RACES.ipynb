### Data cleaning

```python
df[df['Event distance/length'].isin(['50mi','50km'])]
```

```python
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020)]
```

```python
df[df['Event name'] == "Everglades 50 Mile Ultra Run (USA)"]['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2= df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

Drop columns: Athlete club, Athlete Country, Athlete year of birth, Athlete age category

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth', 'Athlete age category'], axis = 1)
```

#### Cleaning missing and duplicated values 

```python
df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna()== 1]
```

```python
df2.dropna(inplace = True)
```

```python
df2.shape
```

```python
df2[df2.duplicated()== True]
```

```python
df2.reset_index(drop= True)
```

```python
df2.dtypes
```

Change data type of `Athlete_age` and `Athlete average speed` columns to int and float type respectively.

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

Rename columns

```python
df2.rename(columns = {'Year of event':'Event_year',
                  'Event dates':'Event_dates',
                  'Event name':'Event_name',
                  'Event distance/length':'Event_distance',
                  'Event number of finishers':'Event_number_of_finishers',
                  'Athlete performance':'Athlete_performance',
                  'Athlete gender':'Athlete_gender',
                  'Athlete average speed':'Athlete_average_speed',
                  'Athlete ID':'Athlete_ID'}, inplace = True)
```

```python
df2.head(5)
```

```python
df3 = df2[['Event_year','Event_dates','Event_name','Event_distance','Event_number_of_finishers','Athlete_ID','Athlete_gender','Athlete_age','Athlete_performance','Athlete_average_speed']]
```

```python
df3.head(5)
```

### Questions on data

Difference in speed in 50km and 50mi between male and female? 

```python
df3.groupby(['Event_distance','Athlete_gender'])['Athlete_average_speed'].mean()
```

What age group are best in the 50m event (20+ races mins)?

```python
df3.query('Event_distance == "50mi"').groupby('Athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

What age group are worst in the 50m event (20+ races mins)?

```python
df3.query('Event_distance == "50mi"').groupby('Athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19')
```

```python
df3['Event_month'] = df3['Event_dates'].str.split('.').str.get(1).astype(int)
```

```python
df3['Event_season'] = df3['Event_month'].apply(lambda x: "Winter" if x > 11 else "Fall" if x > 8 else "Summer" if x > 5 else "Spring" if x > 2 else "Winter")
```

Seasons for the data---> Slower in `Winter` than `Summer`?

```python
df3.groupby('Event_season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

Seasons for the data---> Slower in Winter than Summer (for 50 miler only)?

```python
df3.query("Event_distance =='50mi'").groupby('Event_season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
